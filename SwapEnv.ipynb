```python
import pyarrow as pa 
import pyarrow.parquet as pq 
import numpy as np 
import pandas as pd
from env.DistSwapTradingEnv import DistSwapTradingEnv
import matplotlib.pyplot as plot
import random
```

```python
env = DistSwapTradingEnv(data_file='./data/FRACDIFF_TEST.parquet')
```

```python
prices = [(s*100)+10000 for s in np.sin(time)]
```

```python
env.ind_df = pd.DataFrame({"price":prices}, index=env.ind_df.index[:len(prices)])
```

```python
infos = []
```

```python
pd.DataFrame(infos)[['current_price']].plot(figsize=(30,30))
```
